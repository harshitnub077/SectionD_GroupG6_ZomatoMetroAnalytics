```python

import pandas as pd
import sys
sys.path.insert(0, "..")
from scripts.etl_pipeline import add_kpi_columns

df = pd.read_csv("../data/processed/cleaned_zomato.csv")
df = add_kpi_columns(df)
print(f"Rows: {len(df):,}")
df.head()
```

```python
restaurant_df = df.drop_duplicates(subset=["restaurant_name", "city"])
kpi1 = restaurant_df.groupby("city")["rating_gap"].mean().sort_values().reset_index()
kpi1.columns = ["city", "avg_rating_gap"]
print(kpi1.to_string(index=False))
```

```python
kpi2 = df.groupby("cuisine")["votes"].mean().sort_values(ascending=False).head(15).reset_index()
kpi2.columns = ["cuisine", "avg_votes"]
print(kpi2.to_string(index=False))
```

```python
df.to_csv("../data/processed/tableau_ready_zomato.csv", index=False)
print(f"Saved | Rows: {len(df):,} | Columns: {len(df.columns)}")
print(f"Columns: {df.columns.tolist()}")
```
