```python
import pandas as pd
from scipy import stats

df = pd.read_csv("../data/processed/cleaned_zomato.csv")
restaurant_df = df.drop_duplicates(subset=["restaurant_name", "city"])
print(f"Restaurants: {len(restaurant_df):,}")
```

```python
t_stat, p_value = stats.ttest_rel(restaurant_df["dining_rating"], restaurant_df["delivery_rating"])

print(f"T-statistic : {t_stat:.4f}")
print(f"P-value     : {p_value:.6f}")

if p_value < 0.05:
    print("Result: Significant difference between dining and delivery ratings")
else:
    print("Result: No significant difference")
```

```python
groups = [group["rating_gap"].values for _, group in restaurant_df.groupby("city")]
f_stat, p_value = stats.f_oneway(*groups)

print(f"F-statistic : {f_stat:.4f}")
print(f"P-value     : {p_value:.6f}")

if p_value < 0.05:
    print("Result: Rating gap is significantly different across cities")
else:
    print("Result: No significant difference across cities")
```

```python
bestseller = pd.to_numeric(df[df["best_seller"] == "BESTSELLER"]["votes"], errors="coerce").dropna()
non_seller = pd.to_numeric(df[df["best_seller"].isna()]["votes"], errors="coerce").dropna()

print(f"BESTSELLER count : {len(bestseller)}")
print(f"No tag count     : {len(non_seller)}")

t_stat, p_value = stats.ttest_ind(bestseller, non_seller, equal_var=False)

print(f"Avg votes BESTSELLER : {bestseller.mean():.1f}")
print(f"Avg votes No tag     : {non_seller.mean():.1f}")
print(f"P-value              : {p_value:.6f}")

if p_value < 0.05:
    print("Result: BESTSELLER tag items have significantly more votes")
else:
    print("Result: No significant difference")
```
